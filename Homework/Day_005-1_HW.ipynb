```python
# import library
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

#### 方法一：爬蟲取得 Data

```python
# Crawler of population & countries data
url = 'https://www.worldometers.info/world-population/population-by-country/'
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# Get Population & Countries 
countries = []
population = []
info = []

all_tags = soup.find_all("td")
count = 1
for tag in all_tags:
    t = tag.get_text().strip()
    info.append(t)
    count += 1
    if count == 360:
        break

country_index = 1
population_index = 2
for i in range(30):
    country = info[country_index]
    populations = info[population_index]
    countries.append(country)
    population.append(populations)
    country_index += 12
    population_index += 12
    
    if country_index > len(info):
        break

print(countries)
print(population)
```

```python
# zip two list
list_labels = ['country', 'population']
list_cols = [countries, population]
zipped = zip(list_labels, list_cols)

# Transform to DataFrame
population_data = pd.DataFrame(dict(zipped))
population_data.loc[:,:]
```

#### 方法二：建立模擬Data

```python
countries2 = ['Japan', 'Germany', 'United Kingdom', 'Tanzania', 'Tanzania', 'Japan', 'South Korea']
population2 = np.random.randint(100000, size = 7)

list_labels2 = ['country', 'population']
list_cols2 = [countries2, population2]
zipped2 = zip(list_labels2, list_cols2)
```

```python
# Transform to DataFrame
population_data2 = pd.DataFrame(dict(zipped2))
population_data2.loc[:,:]
```

```python
# Find maximum value of population
countries_pop = population_data2.groupby(by = 'country')['population'].sum()
max_pop = countries_pop.max()
max_pop_df = countries_pop.to_frame() 

max_pop_df.loc[lambda df: df['population'] == max_pop, :]
```
